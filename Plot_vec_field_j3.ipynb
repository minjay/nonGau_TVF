```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io as sci_io
```

```python
data = sci_io.loadmat('vec_field_j3.mat')
```

```python
mag_t = np.sqrt(np.square(data['vec_field_sum_t'][:, 0]) + np.square(data['vec_field_sum_t'][:, 1]))
mag_g = np.sqrt(np.square(data['vec_field_sum_g'][:, 0]) + np.square(data['vec_field_sum_g'][:, 1]))
```

```python
fig, axes = plt.subplots(2, 1, figsize=(15, 15))
```

```python
cmax = np.max([np.max(mag_t), np.max(mag_g)])
```

```python
axes[0].quiver(data['HX'], data['HY'], data['vec_field_sum_t'][:, 0], data['vec_field_sum_t'][:, 1], mag_t, headwidth=2, scale=40, scale_units='inches', clim=[0, cmax])
axes[0].plot(data['xh'][0], data['yh'][0], 'k-')
axes[0].plot(data['xh2'][0], data['yh2'][0], 'k-')
axes[0].set_title(r'j=3, non-Gaussian, $\nu$=2.5', fontsize=20)
axes[0].axis('off');
```

```python
qui = axes[1].quiver(data['HX'], data['HY'], data['vec_field_sum_g'][:, 0], data['vec_field_sum_g'][:, 1], mag_t, headwidth=2, scale=40, scale_units='inches', clim=[0, cmax])
axes[1].plot(data['xh'][0], data['yh'][0], 'k-')
axes[1].plot(data['xh2'][0], data['yh2'][0], 'k-')
axes[1].set_title('j=3, Gaussian', fontsize=20)
axes[1].axis('off');
```

```python
fig.subplots_adjust(right=0.8)
cbar_ax = fig.add_axes([0.85, 0.15, 0.05, 0.7])
fig.colorbar(qui, cax=cbar_ax)
cbar_ax.tick_params(labelsize=20)
```
